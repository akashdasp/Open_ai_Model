```python
enviroment_name='Breakout-v0'
env = gym.make(enviroment_name)
```

```python
episodes=10

for episode in range(1,episodes+1):
    state=env.reset()
    done=False
    score=0
    
    while not done:
        env.render()
        action=env.action_space.sample()
        n_state,reward,done,info=env.step(action)
        score+= reward
    print(episode ,score)
env.close()
```

## Vectorise Envirment  and Train the model

```python
env=make_atari_env('Breakout-v0',n_envs=4,seed=0)
env=VecFrameStack(env,n_stack=4)
```

```python
log_path=os.path.join('traning','logs')
model=A2C('CnnPolicy',env,verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=100000)
```

## Save the Model


```python
a2c_path=os.path.join('traning','saved_model','A2C_breakout_model')
model.save(a2c_path)
```

```python
model=A2C.load(a2c_path,env)
```

### Evalutating the Model


```python
env=make_atari_env('Breakout-v0',n_envs=1,seed=0)
env=VecFrameStack(env,n_stack=4)
```

```python
evaluate_policy(model,env,n_eval_episodes=20,render=True)
```
